```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def sine(x, period=100, offset=0, max=1):
    x = x+offset
    return (np.sin(2*x*np.pi/period)/2 + 0.5) * max

def sawtooth(x, period=100, offset = 0, max=1):
    x = x+offset
    return ((x%period)/period) * max

def exponentialsawtooth(x, period=100, offset = 0, exponent=2, max=1):
    x = x+offset
    return (((x%period)/period) ** exponent) * max

def exponentialsine(x, period=100, offset = 0, exponent=2, max=1):
    x = x+offset
    return ((np.sin(2*x*np.pi/period)/2 + 0.5) ** exponent) * max

def squarewave(x, period=100, offset = 0, ratio=0.5, max=1):
    x = x+offset
    x[x%period > period*ratio] = max
    x[x%period <= period*ratio] = 0
    return x
```

```python
x = np.arange(0,1000,1)
```

```python
y1 = sine(x, 250, offset = -50)
y2 = squarewave(x,200)
y3 = exponentialsawtooth(x, period=200, exponent=0.3, max=0.05)
y4 = exponentialsine(x, period=250, exponent=4, max=1, offset = 100)
```

```python
fig, ax = plt.subplots(2,2,figsize=(12,8))
ax[0][0].plot(x, y1)
ax[0][1].plot(x, y2)
ax[1][0].plot(x, y4)
ax[1][1].plot(x, y3)
```

```python
x = np.arange(0,1000,1)
y = np.arange(0,1000,1)
```

```python
def z(x,y):
    return np.sin(x/100) * np.sin(y/100) * 1.5 ** (x/250) * 1.5 ** (y/250)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
z = z(X,Y)
```

```python
current_generation = np.random.randint(50, 151, size=(100, 2))
```

```python
fig, ax = plt.subplots(1,1,figsize=(12, 12))
ax.imshow(z, cmap='plasma')
ax.axis('off')
ax.scatter(current_generation[:,0], current_generation[:,1], s=5, color='black', alpha=0.5)
```
